```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt 
print(tf.__version__)
tf.config.list_physical_devices('GPU')
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
train_images_norm = train_images / 255.0
test_loss_norm = test_images/255.0
class_names = ['T-shirt/top','Trousers', 'Pullover','Dress','Coat','Sandals','Shirt','Sneakers','Bag','Ankle boot']
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(784, activation='sigmoid'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(784, activation='sigmoid'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(784, activation='sigmoid'),
    tf.keras.layers.Dense(10)    
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(train_images,train_labels, epochs=500)
```

```python
test_loss_test_norm = model.evaluate(test_loss_norm, test_labels, verbose=2)
```
